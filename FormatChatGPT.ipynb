```python
import json
import pandas as pd

DEFAULT_SYSTEM_PROMPT = 'You are an assistant, intelligent. Answer by a accurate, detail the following questions'

def create_dataset(question, answer):
    return {
        "messages": [
            {"role": "system", "content": DEFAULT_SYSTEM_PROMPT},
            {"role": "user", "content": question},
            {"role": "assistant", "content": answer},
        ]
    }
```

```python
if __name__ == "__main__":
    with open("train.jsonl", "w") as f:
        df = pd.read_csv("/content/legaldataset.csv")
        for _, row in df.iterrows():
            example_str = json.dumps(create_dataset(row["question"], row["response"]))
            f.write(example_str + "\n")
```

```python
import json
import random

def reduce_jsonl_file(input_file, output_file, keep_percentage=10):
    # Read all lines from the input JSONL file
    with open(input_file, 'r') as file:
        lines = file.readlines()

    # Calculate the number of lines to keep (10%)
    total_lines = len(lines)
    keep_lines_count = int(total_lines * (keep_percentage / 100.0))

    # Randomly select lines to keep
    lines_to_keep = random.sample(lines, keep_lines_count)

    # Write the selected lines to the output JSONL file
    with open(output_file, 'w') as file:
        for line in lines_to_keep:
            file.write(line)

# Example usage
input_file = '/content/train.jsonl'  # Replace with your input file path
output_file = 'output.jsonl'  # Replace with your desired output file path
reduce_jsonl_file(input_file, output_file)
```
