```python
from safetensors import safe_open

# Replace this with the path to your .safetensors file
file_path = "/content/lora1_khoya.safetensors"

# Open the .safetensors file
with safe_open(file_path, framework="numpy") as f:
    # List all keys (tensor names) in the file
    tensor_keys = f.keys()
    print("Keys in the safetensors file:")
    print(tensor_keys)

    # Display metadata and shape of each tensor
    for key in tensor_keys:
        tensor = f.get_tensor(key)
        print(f"\nKey: {key}")
        print(f"Shape: {tensor.shape}")
        print(f"Dtype: {tensor.dtype}")
```
